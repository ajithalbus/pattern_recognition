```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fle=open('./data_set/HandWritten_data/DATA/FeaturesHW/dA.ldf','r')
arr=[]
i=-1
for line in fle:
    #print line
    i+=1
    if i%3==0 or i%3==1:
        continue
    ln=np.fromstring(line,sep=' ')[1:]
    ln=np.reshape(ln,(ln.shape[0]/2,2))
    
    arr.append(ln)
    
    
```

```python
def curvature(xd,yd,xdd,ydd):
    n=xd*ydd-yd*xdd
    d=(xd**2+yd**2)**1.5
    if d==0:
        if n>0:
            return float('inf')
        elif n<0:
            return float('-inf')
        else:
            return 0
    return n/d
```

```python
j=0
for ass in arr:
    gradx=np.gradient(ass[:,0])
    grady=np.gradient(ass[:,1])
    dgradx=np.gradient(gradx)
    dgrady=np.gradient(grady)
    mn=np.c_[ass,gradx,grady,dgradx,dgrady]

    curv=[curvature(i[2],i[3],i[4],i[5]) for i in mn]

    final=np.c_[mn,curv]
    np.savetxt('./created/dA/dA_'+str(j)+'.txt',final)
    j+=1
```
