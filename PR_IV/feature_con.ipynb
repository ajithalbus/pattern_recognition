```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io as io
```

```python
x=io.loadmat('./data_set/HandWritten_data/3.mat')
v=np.c_[x['xData'][0],x['yData'][0]]
```

```python
print v
    
    
```

```python
def curvature(xd,yd,xdd,ydd):
    n=xd*ydd-yd*xdd
    d=(xd**2+yd**2)**1.5
    if d==0:
        if n>0:
            return float('inf')
        elif n<0:
            return float('-inf')
        else:
            return 0
    return n/d
```

```python
j=0
ass=v
gradx=np.gradient(ass[:,0])
grady=np.gradient(ass[:,1])
dgradx=np.gradient(gradx)
dgrady=np.gradient(grady)
mn=np.c_[ass,gradx,grady,dgradx,dgrady]

curv=[curvature(i[2],i[3],i[4],i[5]) for i in mn]

final=np.c_[mn,curv]
np.savetxt('./created/3_'+str(j)+'.txt',final)
j+=1
```
